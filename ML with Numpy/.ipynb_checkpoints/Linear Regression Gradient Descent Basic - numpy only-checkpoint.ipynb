# Introduction 

We will try to create a machine learning model for Linear Regression by using gradient descent and normal equation. In linear regression problem we want to predict a continous variable which is a Real Number based on our training inputs or so called features and training outputs. This is called a supervised learning since the inputs and outputs for our regression model already given. Generally, given a set of inputs and outputs we want to predict a new outputs based on the new inputs. The hypothesis equation for linear regression for $n$ number of features is:

$$H_{\theta}(x_1,x_2,\ldots,x_n) = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \ldots + \theta_n x_n $$

Let's start it simple if we only has one features our hypothesis equation will be:

$$H_{\theta}(x) = \theta_1 x + \theta_0$$ 

In linear regression problem we want to find $\theta$ is that fit our training inputs and outputs.

As an example we only have one features of X = [1,2,3,4,5] and y = [2,4,6,8,10], what is the suitable value of $\theta_0$ and $\theta_1$ to map $X \rightarrow y$ based on our hypothesis equation. Easily we can figure it out that to get Y we just need to multiply X by 2, hence the value of $\theta = 2$ and $\theta_0 = 0$.


```python
# Define our hypothesis equation
def h(x,theta1,theta0):
    return x*theta1 + theta0
```

```python
X = np.array([1,2,3,4,5]) # Input
y = np.array([2,4,6,8,10]) # Output
plt.scatter(X,y,marker='x',color='red')
plt.plot(X,h(X,2,0))
plt.xlabel('X')
plt.ylabel('y')
```

Given a new input X=24, we can predict it using our hypothesis equation with the parameter of $\theta = 2$ and $\beta=0$ and we get $H(24,2,0)=48$. But in real world cases, our dataset will not be as beautiful as this. There will be a quite a lot of noises and outliers which make our model failed to correctly predict all of our training outputs. Instead we will just need to create a hypothesis equation that can predict outputs such that the error between $H(x)$ and $y$ is minimize. In other word they called it a line of best fit. To calculate the error between our predicted value, $H(x)$ and our true value, $y$ we will use a cost function, $J(\theta)$:

$$ J(\theta) = \frac{1}{2m} \sum_{i=0}^{m} (H_{\theta}(x_i) - y_i)^2$$ 

where m is the total number of our training set. This function usually known as Root Mean Squared Error (RMSE) the inner function in the summation is known as L-2 norm.

## Gradient Descent

The way we update our theta by gradient descent is by using this formula:

$$ \theta_n := \theta_n - \alpha \frac{\partial J(\theta)}{\partial \theta_n}$$

where $\alpha$ is the learning rate. In summary, we want to minimize our $J(\theta)$ by finding the optimized $\theta$ values using gradient technique. By derivation using calculus we get, 

$$\frac{\partial J}{\partial \theta_0} = \frac{1}{m} \sum_{i=0}^{m} (H_{\theta_0}(x_i) - y_i)$$

$$\frac{\partial J}{\partial \theta_1} =  \frac{1}{m} \sum_{i=0}^{m} (H_{\theta_1}(x_i) - y_i) \ x_i$$

$$\theta_0 := \theta_0 - \alpha  \frac{1}{m} \sum_{i=0}^{m} (H_{\theta_0}(x_i) - y_i)$$

$$\theta_1 := \theta_1 - \alpha  \frac{1}{m} \sum_{i=0}^{m} (H_{\theta_1}(x_i) - y_i) \ x_i $$

Update the value simultaniously as shown later in the code.


```python
class LinearRegression:
    
    def __init__(self,X,y,alpha=0.01,iteration=100):
        
        np.random.seed(5)
        self.X = X
        self.y = y.reshape(-1,1)
        self.alpha = alpha
        self.m = len(X)
        self.iteration = iteration
        # Initialize randomly our theta value
        self.theta0 = np.random.randint(1,50)
        self.theta1 = np.random.randint(1,50)
    
    def costFun(self,y_pred):
        
        #Cost Function
        costValue = (1/(2*self.m))*sum(np.power((y_pred - self.y),2))
        
        return costValue
    
    def hypothesisFun(self):
        
        # Hypothesis Function
        return self.X*self.theta1 + self.theta0
    
    def update_theta(self,y_pred):
        
        # Update theta value
        theta0_ = self.theta0 - (self.alpha/self.m)*sum(y_pred-self.y)
        theta1_ = self.theta1 - (self.alpha/self.m)*sum((y_pred-self.y)*self.X)
        
        self.theta0 = theta0_
        self.theta1 = theta1_
        
    def get_theta(self):
        
        return "theta0: {}, theta1 {}".format(self.theta0, self.theta1)
    
      
    def predict(self):
        
        for i in tqdm(range(self.iteration)):        
            y_pred = self.hypothesisFun()
            self.update_theta(y_pred) 

        return y_pred
    
```

```python
clf = LinearRegression(X,y,alpha=0.01,iteration=5000)
```

```python
y_pred = clf.predict()
plt.scatter(X,y,color='r',marker='x')
plt.plot(X,y_pred)
```

```python
clf.get_theta()
```
